# 대학교 데이터

```python
univ_df = pd.read_csv('데이터/스타벅스_대학교 거리 및 인근여부.csv', encoding='euc-kr')
```

```python
univ_df_n = univ_df[['사업장명', '대학거리점수']]
```

```python
cv_store_df_250 = pd.read_csv('데이터/경쟁사수_250m.csv', encoding='euc-kr')
cv_store_df_300 = pd.read_csv('데이터/경쟁사수_300m.csv', encoding='euc-kr')
cv_store_df_500 = pd.read_csv('데이터/경쟁사수_500m.csv', encoding='euc-kr')
```

```python
dataframes = [cv_store_df_250, cv_store_df_300, cv_store_df_500]

for i, df in enumerate(dataframes):
    print('##df:', 'df_' + str(i+1))
    counts = df['기존_저가카페수'].value_counts().sort_index()
    print(counts)
```

```python
dataframes = [cv_store_df_250, cv_store_df_300, cv_store_df_500]

for i, df in enumerate(dataframes):
    print('##df:', 'df_' + str(i+1))
    counts = df['1년이내_저가점포_증가수'].value_counts().sort_index()
    print(counts)
```

```python
dataframes = [cv_store_df_250, cv_store_df_300, cv_store_df_500]

for i, df in enumerate(dataframes):
    print('##df:', 'df_' + str(i+1))
    counts = df['기존_중고가카페수'].value_counts().sort_index()
    print(counts)
```

```python
dataframes = [cv_store_df_250, cv_store_df_300, cv_store_df_500]

for i, df in enumerate(dataframes):
    print('##df:', 'df_' + str(i+1))
    counts = df['1년이내_중고가점포_증가수'].value_counts().sort_index()
    print(counts)
```

```python
# 병합
merge_df = pd.merge(base_df, univ_df_n, on='사업장명', how='inner') # 스타벅스 + 대학 정보
merge_df = pd.merge(merge_df, sub_df_n, on='사업장명', how='inner') # + 지하철 정보
merge_df = pd.merge(merge_df, cv_store_df_500, on='사업장명', how='inner') # + 경쟁사 정보
merge_df = pd.merge(merge_df, price_df_n, on='사업장명', how='inner') # + 공시지가 정보
merge_df = pd.merge(merge_df, liv_pop_df, on='사업장명', how='inner') # + 생활인구 정보
merge_df
```

```python
merge_df.rename(columns={'1년이내_저가점포_증가수':'1년이내_저가카페_증가수', 
                         '1년이내_중고가점포_증가수': '1년이내_중고가카페_증가수'}, inplace=True)
```

```python
merge_df_n = merge_df.dropna()
```
